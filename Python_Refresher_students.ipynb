# 1. Data Types and Variables

Python has several basic data types that are fundamental to working with data:
- **int**: Integers (whole numbers)
- **float**: Floating-point numbers (decimals)
- **str**: Strings (text)
- **bool**: Boolean (True/False)
- **list**: Ordered collection of items
- **dict**: Unordered collection of key-value pairs

__`Step 1`__ Create variables of different types and check their data type using the `type()` function.

```python
# Integer
int_value = 42
print(type(int_value))  # Output: <class 'int'>
# Float
float_value = 3.14
print(type(float_value))  # Output: <class 'float'>
# String
string_value = "Hello, World!"
print(type(string_value))  # Output: <class 'str'>
# Boolean
boolean_value = True
```

# 2. Lists and Basic Operations

Lists are ordered collections that can hold multiple items. They are mutable (can be changed) and are indexed starting from 0.

__`Step 2`__ Create a list and access elements using indexing.

```python
# Create a list of features
my_list= [1, 2, 3, 4, 5, 5]
# Access elements by index
print(my_list[0])  # Output: 1
# Get length of list
print(len(my_list))  # Output: 5
```

```python
# Start with a list

# Append (add to end)
my_list.append(6)
# Insert at specific position
my_list.insert(0, 0)  # Insert 0 at index 0
# Remove specific item
my_list.remove(3)  # Remove the first occurrence of 3
```

# 3. Dictionaries

Dictionaries store data as key-value pairs. They are very useful for organizing related data.

__`Step 4`__ Create and manipulate dictionaries.

```python
# Create a dictionary
my_dictioniary = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}

# Access values using keys
my_dictioniary["name"]  # Output: 'Alice'
my_dictioniary.get("age")  # Output: 30

# Add new key-value pair
my_dictioniary["email"] = "alice@example.com"

# Get all keys and values
keys = my_dictioniary.keys()  # Output: dict_keys(['name', 'age', 'city', 'email'])
values = my_dictioniary.values()  # Output: dict_values(['Alice', 30, 'New York', 'alice@example.com'])
items = my_dictioniary.items()  # Output: dict_items([('name', 'Alice'), ('age', 30), ('city', 'New York'), ('email', '
print(items)  # Output: dict_items([('name', 'Alice'), ('age', 30), ('city', 'New York'), ('email', '
```

# 4. Control Flow: Conditionals (if/else)

Conditionals allow us to execute different code based on conditions.

__`Step 5`__ Use if/else statements to make decisions.

```python
age = 30
if age < 18:
    print("no beer for you")
elif age < 21:
    print("have a beer, but only light beer")
else:
    print("have a beer")
```

# 5. Loops: for and while

Loops allow us to repeat code multiple times.

__`Step 6`__ Use for loops to iterate over lists.

```python
# For loop over a list

for item in my_list:
    print(item)

# For loop with index
for i in range(len(my_list)):
    print(f"Index: {i}, Value: {my_list[i]}")
```

__`Step 7`__ Use while loops for conditional repetition.

```python
# While loop
import random
random_number = 0
security_counter = 10000
while random_number != 5:
    print(f"Random Number: {random_number}")
    random_number = random.randint(6, 10)
    security_counter -= 1
    if security_counter <= 0:
        print("Security counter reached zero, breaking loop to prevent infinite execution.")
        break
```

# 6. Functions

Functions allow us to reuse code and make it more organized.

__`Step 8`__ Define and call functions with parameters and return values.

```python
# Function with parameters and return value
def my_function(x):
    return x * 2

# Call the function
result = my_function(5)
print(result)  # Output: 15
```

__`Step 8a`__ Define functions without parameters.


```python
# Function without parameters
def no_param_function():
    print(my_list)
# Call the function
no_param_function()  # Output: [0, 1, 2, 4, 5, 5, 6]
```

__`Step 8b`__ Define functions with multiple parameters.


```python
# Function with multiple parameters
def multi_param_function(x: int, y: int) -> int:
    return x + y

# Call with positional arguments

multi_param_function(5, 10)  # Output: 15

# Call with keyword arguments
multi_param_function(x=5, y=10)  # Output: 15
```

__`Step 8c`__ Define functions that return multiple values.


```python
# Function returning multiple values
my_tuple = (1, "hello", 3.14)
def return_multiple() -> tuple:
    return my_tuple[1], my_tuple[2]
# Call the function and unpack return values
values = return_multiple()
len(values)  # Output: <class 'tuple'>
```

__`Step 8d`__ Define nested functions (functions inside functions).


```python
# Function with nested function
def big_function(x):
    def nested_function_1(y):
        return y * 2
    
    def nested_function_2(z):
        return z + 5
    
    if x < 10:
        return nested_function_1(x)
    else:
        return nested_function_2(x)

print(big_function(5))  # Output: 10 (5 * 2)
print(big_function(15))  # Output: 20 (15 + 5)
# Create a function that doubles values
def double_value(x):
    return x * 2
```

# 7. String Operations

# 10. NumPy Basics

NumPy provides efficient numerical operations on arrays.

__`Step 12`__ Use NumPy arrays for mathematical operations.

```python
import numpy as np
my_list_2 = [2.5,3,4,5] 
# Create arrays
my_np_array_1 = np.array(my_list_2, dtype=int)  # This will create a NumPy array from the list of squares
my_np_array_2 = np.ones(4, dtype=int)  # Create an array of ones with length 5 
print(my_np_array_2)  # Output: [2 3 4 5]

print(my_np_array_1)  # Output: [2 3 4 5]
# Element-wise operations

print(my_np_array_2 + my_np_array_1)  # Output: [3 4 5 6 6]
print(my_np_array_2 * my_np_array_1)  # Output: [2 3 4 5 5]
print(my_np_array_2.dot(my_np_array_1))  # Output: [2 3 4 5 5]

# Statistical operations\
mean_value = my_np_array_1.mean()  # Output: 3.5
median_value = np.median(my_np_array_1)  # Output: 3.5
std_dev = my_np_array_1.std()  # Output: 1.118
print(f"Mean: {mean_value}, Median: {median_value}, Standard Deviation: {std_dev}")
```

# 11. Pandas DataFrame Operations

Pandas DataFrames are the primary data structure for data science in Python.

__`Step 13`__ Perform common DataFrame operations.

```python
# Load the data
type(df)  # Output: <class 'pandas.core.frame.DataFrame'>
# Select single column
# difference beween loc and iloc
df.loc[:, 'BD1']  # Output: Series with 'BD1' column
df.iloc[:, 0]  # Output: Series with the first column (BD1 in this DataFrame)
# Select multiple columns
df.loc[:, ['BD1', 'BD2', 'DrugPlant']]


# Filter rows
df[df['BD1'] > 100]  # Output: DataFrame with
# whn I want them equal to something the sign is == not >
df[df['BD1'] == 100]  # Output: DataFrame with rows where 'BD1' is equal to 100
```

__`Step 14`__ Use groupby and aggregation functions.

```python
# Group by and calculate mean

df.groupby('DrugPlant')['BD1'].mean()  # Output: Series with mean 'BD1' for each 'DrugPlant' category
# Count occurrences
df['DrugPlant'].value_counts()  # Output: Series with count of each unique value in 'DrugPlant' column

# Multiple aggregations
df.groupby('DrugPlant')['BD1'].agg(['mean', 'median', 'std'])  # Output: DataFrame with mean, median, and std of 'BD1' for each 'DrugPlant' category
```

# 12. Exception Handling

Exception handling allows us to gracefully handle errors.

__`Step 15`__ Use try/except blocks to handle errors.

```python
# Example 1: Division by zero

try:
    result = 10 / 0

except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
    result = 10 / 1

#Example 2: Key error
data = {'name': 'John', 'age': 25}
try:
    print(data['email'])  # This key doesn't exist
except KeyError:
    print("Error: Key not found.")
    data['email'] = "Email not found"

# Example 3: General exception
values = [1, 2, 3]
try:
    print(values[10])  # Index out of range
except IndexError:
    print("Error: Index out of range.") 
    print(values[-1])
```
